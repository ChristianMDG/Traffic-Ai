## 1. Import des librairies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Chargement des données

```python
df = pd.read_csv("../data/raw/Metro_Interstate_Traffic_Volume.csv")

df.head()
```

## 4. Statistiques descriptives

## 5. Valeurs manquantes

```python
df.isnull().sum()
```

## 6. Analyse de la variable cible

```python
plt.figure(figsize=(10,5))
df["traffic_volume"].hist(bins=50)
plt.title("Distribution du volume de trafic")
plt.xlabel("Traffic Volume")
plt.ylabel("Frequency")
plt.show()
```

## 7. Analyse des variables catégorielles

```python
plt.figure(figsize=(8,4))
df['weather_main'].value_counts().plot(kind='bar')
plt.title("Répartition des conditions météo")
plt.xlabel("Weather")
plt.ylabel("Count")
plt.show()
```

# Nettoyage des données

## Objectif
Préparer le dataset pour le feature engineering et l’entraînement du modèle. 
- Convertir `date_time` en datetime
- Créer des colonnes temporelles utiles
- Traiter les valeurs manquantes
- Vérifier types et incohérences

## 1. Import des librairies

```python
import pandas as pd
import numpy as np
```

## 2. Chargement des données

```python
df = pd.read_csv("../data/raw/Metro_Interstate_Traffic_Volume.csv")
df.head()
```

## 3. Conversion de date_time en datetime

```python

df['date_time'] = pd.to_datetime(df['date_time'])
```

## 4. Création de colonnes temporelles

```python
df['hour'] = df['date_time'].dt.hour
df['day'] = df['date_time'].dt.day
df['month'] = df['date_time'].dt.month
df['weekday'] = df['date_time'].dt.weekday
df['is_weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)


df.head()
```

## 5. Vérification des valeurs manquantes

```python
df = df[df['traffic_volume'] >= 0].copy()
df.fillna(method='ffill', inplace=True)  
```

## 6. Vérification des incohérences

```python

(df['traffic_volume'] < 0).sum()
```

## 7. Sauvegarde des données nettoyées

```python
df.to_csv("../data/processed/traffic_cleaned.csv", index=False)
```

## Conclusion

- `date_time` converti en datetime
- Nouvelles colonnes temporelles créées : hour, day, month, weekday, is_weekend
- Valeurs manquantes traitées
- Dataset prêt pour le feature engineering

# Feature Engineering

## Objectif
Créer des nouvelles features à partir du dataset nettoyé pour améliorer la performance du modèle.

## 1. Import et chargement des données

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

df = pd.read_csv("../data/processed/traffic_cleaned.csv")
df.head()
```

## 2. Moyennes glissantes du trafic

```python
df = df.sort_values('date_time')
df['traffic_rolling_3'] = df['traffic_volume'].rolling(window=3, min_periods=1).mean()
df['traffic_rolling_6'] = df['traffic_volume'].rolling(window=6, min_periods=1).mean()
```

## 3. Encodage des variables catégorielles

```python

df = df.drop(columns=['date_time'], errors='ignore')

cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Colonnes catégorielles à encoder :", cat_cols)

df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

## 4. Création de variables temporelles avancées

```python
df['part_of_day'] = df['hour'].apply(lambda x: 
                                     'morning' if 6 <= x < 12 else 
                                     'afternoon' if 12 <= x < 18 else 
                                     'evening' if 18 <= x < 24 else 'night')
```

## 5. Sauvegarde du dataset final pour modélisatio

```python
df.to_csv("../data/processed/traffic_features.csv", index=False)
```
